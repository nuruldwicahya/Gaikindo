```python
gaikindo_df['TOTAL_IMPORT'] = gaikindo_df[month_cols].sum(axis=1)
```

```python
# Hitung ulang total dan pertumbuhan
monthly_total = gaikindo_df[month_cols].sum()
monthly_df = monthly_total.reset_index()
monthly_df.columns = ['Month', 'Total_Import']
monthly_df['MoM_Growth (%)'] = monthly_df['Total_Import'].pct_change() * 100

# Set tema background putih
sns.set_theme(style="whitegrid")

# Plot line chart
plt.figure(figsize=(10, 5))
ax = sns.lineplot(data=monthly_df, x='Month', y='Total_Import', marker='o', linewidth=2.5)

# Judul dan label
plt.title('Total Impor Tertinggi Bulan April dan Turun Drastis Bulan May')
plt.ylabel('Unit Kendaraan')

# Format angka di Y-axis agar tidak pakai 1e7
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Tambahkan label angka di titik chart (warna putih)
for i, row in monthly_df.iterrows():
    plt.text(i, row['Total_Import'], f"{int(row['Total_Import']):,}", 
             color='black', fontsize=9, fontweight='bold',
             ha='center', va='bottom')

# Tampilkan grid dan rapikan layout
plt.grid(True)
plt.tight_layout(pad=2.5)  # tambahkan ruang ekstra
plt.title("Total Impor Tertinggi Bulan April dan Turun Drastis Bulan May", pad=20)
plt.savefig('dashboard_impor_kendaraan.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
ax = sns.lineplot(data=monthly_df, x='Month', y='MoM_Growth (%)',
                  marker='o', linewidth=2.5, color='orange')

plt.axhline(0, color='gray', linestyle='--')

# Judul dan label
plt.title('Month Growth Signifikan Pada Bulan Juni')
plt.ylabel('Pertumbuhan (%)')

# Format Y agar tampil angka biasa
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:.1f}%'))

# Tambahkan label angka (warna putih)
for i, row in monthly_df.iterrows():
    plt.text(i, row['MoM_Growth (%)'], f"{row['MoM_Growth (%)']:.1f}%",
             color='black', fontsize=9, fontweight='bold',
             ha='center', va='bottom')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
bulan_cols = ['JANUARY', 'FEBRUARY', 'MARCH', 'APRIL', 'MAY', 'JUNE']
gaikindo_df[bulan_cols] = gaikindo_df[bulan_cols].apply(pd.to_numeric, errors='coerce')

# Buat pivot table
brand_monthly = gaikindo_df.pivot_table(
    index='BRAND',
    values=bulan_cols,
    aggfunc='sum'
)

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(brand_monthly, cmap='YlGnBu', norm=LogNorm(), linewidths=0.5, linecolor='gray')
plt.title('Heatmap Volume Impor Berdasarkan Brand (Jan–Jun)', fontsize=14)
plt.xlabel('Bulan')
plt.ylabel('Brand')
plt.tight_layout()
plt.show()
```

```python
country_total = gaikindo_df.groupby('ORIGIN COUNTRY')[month_cols].sum().sum(axis=1).sort_values(ascending=False)
```

```python
country_total_nonzero = country_total[country_total > 0]

plt.figure(figsize=(12, 8))
sns.barplot(x=country_total_nonzero.values, 
            y=country_total_nonzero.index, 
            palette='magma')

# Pakai skala log agar semua bar terlihat
plt.xscale('log')

# Format angka agar tidak dalam notasi ilmiah
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Tambahkan label angka
max_val = max(country_total_nonzero.values)

for i, (val, label) in enumerate(zip(country_total_nonzero.values, country_total_nonzero.index)):
    if val > 0.15 * max_val:
        # label putih di dalam bar
        plt.text(val * 0.98, i, f'{int(val):,}', color='white', va='center', ha='right', fontsize=9)
    else:
        # label hitam di luar bar
        plt.text(val * 1.1, i, f'{int(val):,}', color='black', va='center', ha='left', fontsize=9)

plt.title('Total Impor Berdasarkan Negara Asal (Jan–Jun 2025)')
plt.xlabel('Unit Kendaraan')
plt.ylabel('ORIGIN COUNTRY')
plt.tight_layout()
plt.show()
```

```python
months = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN"]
values = [9298340, 8439442, 7802013, 9987035, 5162468, 9315528]
growth = [0] + [((values[i] - values[i-1]) / values[i-1]) * 100 for i in range(1, len(values))]

# Dummy data (ganti pakai data asli kamu di notebook)
brands = ["Honda", "Toyota", "Mitsubishi", "Suzuki", "Nissan"]
brand_values = [45000, 38000, 25000, 20000, 15000]
countries = ["Japan", "Thailand", "China", "Korea", "Others"]
country_values = [50000, 30000, 20000, 15000, 8000]
types = ["Non-Elektrik", "Elektrik"]
market_share = [85000, 15000]
importers = ["Honda Prospect", "Toyota Astra", "Mitsubishi Motor", "Suzuki Indomobil", "Nissan Motor"]
import_values = [60000, 40000, 30000, 20000, 15000]

fig, axes = plt.subplots(3, 2, figsize=(20, 18))
axes = axes.flatten()

# 1. Total impor tertinggi bulan April & turun drastis Mei
axes[0].plot(months, values, marker='o', color='steelblue')
axes[0].set_title("1. Impor Tertinggi April & Turun Drastis Mei")
axes[0].scatter(["APR"], [max(values)], color='green', s=120, label="Tertinggi")
axes[0].scatter(["MAY"], [min(values)], color='red', s=120, label="Terendah")
axes[0].legend()

# 2. Month growth signifikan pada bulan Juni
axes[1].plot(months, growth, marker='o', color='orange', linewidth=2)
axes[1].axhline(0, color='black', linestyle="--")
axes[1].set_title("2. Pertumbuhan Bulanan – Lonjakan di Juni")
for i, g in enumerate(growth):
    axes[1].text(i, g + (2 if g>=0 else -2), f"{g:.1f}%", ha='center')

# 3. Minat customer tertinggi pada brand Honda
axes[2].bar(brands, brand_values, color=['green' if b=="Honda" else 'skyblue' for b in brands])
axes[2].set_title("3. Minat Customer Tertinggi: Honda")

# 4. Total impor per Negara Asal
axes[3].pie(country_values, labels=countries, autopct='%1.1f%%', startangle=140)
axes[3].set_title("4. Total Impor per Negara Asal")

# 5. Dominasi pasar oleh kendaraan Non Elektrik
axes[4].bar(types, market_share, color=['steelblue', 'limegreen'])
axes[4].set_title("5. Dominasi Pasar: Non Elektrik")

# 6. Importir kendaraan tertinggi: Honda Prospect Motor
axes[5].bar(importers, import_values, color=['green' if i==0 else 'skyblue' for i in range(len(importers))])
axes[5].set_title("6. Importir Kendaraan Tertinggi")

# Styling
for ax in axes:
    ax.grid(True, linestyle="--", alpha=0.5)
    for label in ax.get_xticklabels():
        label.set_rotation(20)

plt.tight_layout(pad=4)
plt.show()
```
